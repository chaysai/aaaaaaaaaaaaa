```python
import pandas as pd
```

```python
data=pd.read_csv("telecom_customer_churn.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data=data.drop(["Churn Category","Churn Reason"],axis=1)
```

```python
data.columns
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64','uint8']
odf = data.select_dtypes(exclude=numerics)
for i in odf:
    print(i," ",data[i].unique())
```

```python
data = data.drop(columns=["Customer ID","City","Zip Code","Latitude","Longitude"],axis=1)
```

```python
data.info()
```

```python
col=["Avg Monthly Long Distance Charges","Avg Monthly GB Download"]
data[col]=data[col].fillna(0)
```

```python
data.info()
```

```python
data=data.dropna()
```

```python
data.info()
```

```python
for i in data.select_dtypes(include="object"):
    if(len(data[i].unique())<3):
        state = pd.get_dummies(data[i],drop_first=True,prefix=i)
        data = pd.concat([data,state],axis=1)
        data = data.drop([i],axis=1)
    else:
        state = pd.get_dummies(data[i],prefix=i)
        data = pd.concat([data,state],axis=1)
        data = data.drop([i],axis=1)
data.info()
```

```python
x=data.drop("Customer Status", axis =1).values
y=data.loc[:,"Customer Status"].values
print(x[0])
print(y)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.30, random_state=5)
```

```python
from sklearn.preprocessing import StandardScaler

ss=StandardScaler()

x_train=ss.fit_transform(x_train)
x_test=ss.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
